# Matrix Plot 
Create Matrix plots especially heatmap

```python
import seaborn as sns
%matplotlib inline
tips=sns.load_dataset('tips')
flights=sns.load_dataset('flights')
```

```python
tips.head()
```

```python
flights.head(30)
```

```python
tc=tips.corr()
tc
```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```python
tips
tips.pivot_table(index='day',columns='total_bill',values='tip')
```

```python
flights
fp=flights.pivot_table(index='month',columns='year',values='passengers')
fp
```

```python
flights.corr()
```

```python
sns.heatmap(fp,cmap='coolwarm',linecolor='white',linewidths=3)
```

```python
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
```

# Grid

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
iris=sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
sns.PairGrid(iris)
```

```python
g=sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
tips=sns.load_dataset('tips')
tips.head()
new_g=sns.FacetGrid(data=tips,col='time',row='smoker')
new_g.map(sns.distplot,'total_bill')
```

```python
new_g.map(plt.scatter,'total_bill','tip')
```

# Regression Plot

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',row='time',hue='sex',
           aspect=1,size=8,scatter_kws={'s':100},markers=['o','v'])
```
