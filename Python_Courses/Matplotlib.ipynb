```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
x
```

```python
y=x**2.
y
```

# Functional 

```python
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('X-Y')
# plt.show()
```

# Create multiple plot 

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('left-side')

plt.subplot(1,2,2)
plt.plot(y,x)
plt.ylabel('Y-Axis')
plt.title('right-side')


```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
x
```

```python
y=x**2
y
```

```python
plt.plot(x,y,color='r')
plt.xlabel('x-data',color='r')
plt.ylabel('y-data',color='g')
plt.title('results')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.xlabel('x-data',color='r')
plt.ylabel('y-data',color='g')
plt.title('results of plot_one')

plt.subplot(1,2,2)
plt.plot(x,y)
plt.xlabel('x-data',color='r')
plt.ylabel('y-data',color='g')
plt.title('results of plot_two')
```

# object orientent methods

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_title("ajdsjfa")
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.6,0.2,0.2,0.6])
axes1.plot()
axes2.plot()
```

# Matplotlib Part2

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
print(fig,axes)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=3)
print(fig,axes)
```

```python
fig,axes3=plt.subplots(nrows=2,ncols=2)
# [0][0].plot(x,y)
# plt.tight_layout()
fig.subplots_adjust(hspace=1)
fig.subplots_adjust(wspace=1)
axes3[0][0].plot(x,y)
axes3[0][0].set_xlabel('xlabel')
axes3[0][0].set_ylabel('xlabel')
axes3[0][0].set_title('xlabel')
axes3[0][1].plot(x,y)
axes3[0][1].set_xlabel('xlabel')
axes3[0][1].set_ylabel('xlabel')
axes3[0][1].set_title('xlabel')
axes3[1][0].plot(x,y)
axes3[1][0].set_xlabel('xlabel')
axes3[1][0].set_ylabel('xlabel')
axes3[1][0].set_title('xlabel')
axes3[1][1].plot(y,x)
axes3[1][1].set_xlabel('xlabel')
axes3[1][1].set_ylabel('xlabel')
axes3[1][1].set_title('xlabel')
```
