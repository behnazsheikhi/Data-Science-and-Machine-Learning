```python
from sklearn.datasets import  make_blobs
```

```python
data=make_blobs(n_samples=200,n_features=2,random_state=101,centers=4,cluster_std=1.8)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=4)
```

```python
kmeans
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
fig,(x1,x2)=plt.subplots(1,2,sharey=True,figsize=(10,6))
x1.set_title('K Means')
x1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')

x2.set_title('Original Labels')
x2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```
