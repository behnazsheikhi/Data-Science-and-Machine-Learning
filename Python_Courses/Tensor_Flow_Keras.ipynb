# We have to normalize the data 
when we work with neural network(weights and biases) and we have large data, error may happen. In order to avoid this 
we normalize the data

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
# Fit claculate the parameters to perform the actual scaler later on 
# Calculate the std, the min and max
# MinMaxScaler depend on the std and min and max
```

```python
X_train=scaler.transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
# X_train.max()
# After fit and transform the data, the data on X_train and X_test is between 0 to 1 -> min 0 and max 1
```

# Part Two

# Creating the model, Generating the model, Evaluate prformance of the model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
help(Sequential)
```

```python
help(Dense)
```

```python
model=Sequential([Dense(4,activation='relu'),
                 Dense(4,activation='relu'),
                 Dense(4,activation='relu')],
                Dense(1))
```

```python
model=Sequential()
```

```python
model.add(Dense(4,activation='relu'))
model.add(Dense(4,activation='relu'))
model.add(Dense(4,activation='relu'))
model.add(Dense(1))
model.compile(optimizer='rmsprop',loss='mse')
```

```python
model.fit(X_train,y_train,epochs=250)
```

```python
model.history.history
```

```python
loss_data=pd.DataFrame(model.history.history)
```

```python
loss_data
```

```python
loss_data.plot()
```
