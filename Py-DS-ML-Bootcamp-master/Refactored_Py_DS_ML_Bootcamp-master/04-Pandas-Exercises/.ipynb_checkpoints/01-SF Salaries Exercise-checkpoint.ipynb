** How many people have the word Chief in their job title? (This is pretty tricky) **

```python
sal.head()
```

```python
sal[sal["JobTitle"].str.contains("Chief")]
```

```python
sal['JobTitle']
```

```python
def chief_string(tilte):
    
    if 'chief' in tilte.lower().split():
        return True
    else:
        return False
    
```

```python
chief_string('CAPTAIN III (POLICE DEPARTMENT)')
```

```python
sum(sal['JobTitle'].apply(lambda x:chief_string(x)))
```
